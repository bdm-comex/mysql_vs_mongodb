```python
def sql_select_thread_function(name):
    #logging.info("Thread %s: starting", name)
    #time.sleep(2)
    


    try:
        host_var = ''
        port = ''
        if(use_proxy):
            host_var = '10.158.0.2'
            port_var = '6033'
        else:
            host = '10.115.96.3'
            port = '3306'
        
        connection = mysql.connector.connect(host=host_var,
                                             port=port_var,
                                             database='projeto',
                                             user='root',
                                             password='',
                                             charset='utf8')
        #connection = mysql.connector.connect(host='35.199.120.151',
        #                                     port='3306',
        #                                     database='projeto',
        #                                     user='root',
        #                                     password='',
        #                                     charset='utf8')        
        if connection.is_connected():
            db_Info = connection.get_server_info()

            db_cursor = connection.cursor()




            co_ano = str(random.randint(1997, 2020))
            co_mes = str(random.randint(1, 12))
            if random.randint(1,10) % 2 == 0:
                fluxo = "EXP"
            else:
                fluxo = "IMP"

            sql = 'SELECT * \
                   FROM projeto.base_dados \
                   WHERE co_ano = '+co_ano+' and \
                         co_mes = '+co_mes+' and \
                         fluxo = "'+fluxo+'" \
                   ORDER BY vl_fob'
            
#            sql = 'SELECT co_ano, co_mes, fluxo, SUM(vl_fob) AS valor\
#            FROM projeto.base_dados\
#            WHERE co_ano = 2020 and co_mes = 3 and fluxo = "EXP"\
#            GROUP BY co_ano, co_mes'


            if (print_query): print(sql + '\n')

            if (execute_query):
                db_cursor.execute(sql)

                if (print_data):
                    table_rows = db_cursor.fetchall()
                    df = pd.DataFrame(table_rows)
                    print(df)


            #connection.commit()
            connection.close()
            db_cursor.close()

    except Error as e:
        print("Error while connecting to MySQL", e)
    finally:
        if (connection.is_connected()):
            db_cursor.close()
            connection.close()
            print("MySQL connection is closed")


    #logging.info("Thread %s: finishing", name)
```

```python
#teste automatizado sem log
workers_list = [256, 512 ]


print("starting sql select multi thread ")

for i in range(len(workers_list)):
    
    workers_var = workers_list[i]

    rounds = 1
    
    for r in range(0, rounds):


        workers_var = workers_var
        max_workers_var = 8


        start = time.time()

        if __name__ == "__main__":
            format = "%(asctime)s: %(message)s"
            logging.basicConfig(format=format, level=logging.INFO, datefmt="%H:%M:%S")

            with concurrent.futures.ThreadPoolExecutor(max_workers=max_workers_var) as executor:
                executor.map(sql_select_thread_function, range(workers_var))


        end = time.time()
        print("finishing...sql select. Clients: " + str(workers_var) + ", Pool: "+str(max_workers_var))
        print(">>>round ("+str(r)+"), execution time: " + str(end - start) )
    
```
