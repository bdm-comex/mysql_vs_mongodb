```python
#Monta série histórica mensal da balança comercial (resultado da diferença entre o valor das exportações e o das importações)
import string
import random
import pandas as pd
import time
from pymongo import MongoClient
from pprint import pprint
import traceback



execute_query=True
print_query=True
print_data=False


MONGO_URI = 'mongodb://root:eLym6ybM36h8@10.150.0.2:26000'

print("starting mongo select")
print("série histórica mensal da balança comercial (resultado da diferença entre o valor das exportações e o das importações)" )

start = 0




try:
    
    client = MongoClient(MONGO_URI)
    db=client.comex
    

    start = time.time()
    seriehist = db.base_dados.aggregate(
      [
      {"$group": 
        {
           "_id": ["$CO_ANO","$CO_MES"],
           "saldo":  {"$sum": {"$cond": [ { "$eq": [ "$FLUXO", "EXP" ] }, "$VL_FOB", {"$multiply": [ -1, "$VL_FOB" ]}]}}
        }
      },
      {"$sort":  { "_id":1}
      }
      ])   
    
    end = time.time()
    print(">>>execution time: " + str(end - start) )    
    
    if(print_data):
        for item in seriehist:
            print(item)




    client.close()
    
except Exception as e:
    print("Exception seen:", e)
    #traceback.print_exc(file = sys.stdout)
finally:
    client.close()
```

```python
# exportações de EUA, China e Argentina e calcula a participação de cada uma delas nas exportações totais

import string
import random
import pandas as pd
import time
from pymongo import MongoClient
from pprint import pprint
import traceback



execute_query=True
print_query=False
print_data=False


MONGO_URI = 'mongodb://root:eLym6ybM36h8@10.150.0.2:26000'

print("starting mongo select")
print("# exportações de EUA, China e Argentina e calcula a participação de cada uma delas nas exportações totais" )

start = 0



try:
    
    client = MongoClient(MONGO_URI)
    db=client.comex
    
    start = time.time()
    seriehist = db.base_dados.aggregate( 
     [
      { "$match": { "FLUXO": "EXP"} },
      { "$group": {
           "_id": {"CO_ANO": "$CO_ANO", "CO_MES": "$CO_MES"},
           "valor_exp_mes":  {"$sum": "$VL_FOB"},
           "valores_exp_pais_mes": { "$push": {"CO_PAIS": "$CO_PAIS", "VL_FOB": "$VL_FOB"}}
        }
      },
      { "$unwind" : "$valores_exp_pais_mes" },
      { "$match": { "valores_exp_pais_mes.CO_PAIS" :{"$in":[ 249, 160, 63] }} },      
      { "$group": {
           "_id": {"CO_ANO": "$_id.CO_ANO", "CO_MES": "$_id.CO_MES", "CO_PAIS": "$valores_exp_pais_mes.CO_PAIS"},
           "total_exp_mes": {"$first": "$valor_exp_mes"},
           "exp_pais_mes":  {"$sum": "$valores_exp_pais_mes.VL_FOB"},
        }
      },
      { "$project": {
           "exp_pais_mes": 1,
           "total_exp_mes": 1,
           "participacao": {"$multiply": [{"$divide": ["$exp_pais_mes","$total_exp_mes"] }, 100]}
        }
      },
      { "$sort" : { "_id": 1 } }
    ],
    allowDiskUse=True )    

    end = time.time()
    print(">>>execution time: " + str(end - start) )    
    
    if(print_data):
        for item in seriehist:
            print(item)




    client.close()
    
except Exception as e:
    print("Exception seen:", e)
    #traceback.print_exc(file = sys.stdout)
finally:
    client.close()
```
